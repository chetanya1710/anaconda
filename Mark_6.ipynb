```python
retail_new_clean_cluster1=pd.read_csv('retail_new_clean_cluster1.csv')
```

```python
retail_new_clean_cluster1.head()
```

```python
Invoice_vs_Description_matrix_cluster1=(retail_new_clean_cluster1.groupby(['Description'])
                                            .apply(lambda x: {y:1 for y in x['InvoiceNo']})
                                        .apply(pd.Series)
                                        .fillna(0)
                                        .reset_index())
Invoice_vs_Description_matrix_cluster1=Invoice_vs_Description_matrix_cluster1.set_index('Description')
```

```python
Invoice_vs_Description_matrix_cluster1.head()
```

```python
Description_vs_Invoice_matrix_cluster1=(retail_new_clean_cluster1.groupby(['InvoiceNo'])
                                            .apply(lambda x: {y:1 for y in x['Description']})
                                        .apply(pd.Series)
                                        .fillna(0)
                                        .reset_index())
Description_vs_Invoice_matrix_cluster1=Description_vs_Invoice_matrix_cluster1.set_index('InvoiceNo')
```

```python
Description_vs_Invoice_matrix_cluster1.head()
```

```python
user_user_jac_sim = 1-pairwise_distances(Invoice_vs_Description_matrix_cluster1.T, metric = "jaccard")
user_user_jac_sim = pd.DataFrame(user_user_jac_sim, index=Invoice_vs_Description_matrix_cluster1.columns, columns=Invoice_vs_Description_matrix_cluster1.columns)
user_user_jac_sim.head()
```

```python
user_user_cor_sim = 1-pairwise_distances(Invoice_vs_Description_matrix_cluster1.T, metric = "correlation")
user_user_cor_sim = pd.DataFrame(user_user_cor_sim, index=Invoice_vs_Description_matrix_cluster1.columns, columns=Invoice_vs_Description_matrix_cluster1.columns)
user_user_cor_sim.head()
```

```python
user_user_euc_sim = 1-pairwise_distances(Invoice_vs_Description_matrix_cluster1.T, metric = "euclidean")
user_user_euc_sim = pd.DataFrame(user_user_jac_sim, index=Invoice_vs_Description_matrix_cluster1.columns, columns=Invoice_vs_Description_matrix_cluster1.columns)
user_user_euc_sim.head()
```

```python
item_item_corr_sim = 1-pairwise_distances(Description_vs_Invoice_matrix_cluster1.T, metric = "correlation")
item_item_corr_sim = pd.DataFrame(item_item_corr_sim, index=Description_vs_Invoice_matrix_cluster1.columns, columns=Description_vs_Invoice_matrix_cluster1.columns)
item_item_corr_sim.head()
```

```python
item_item_jac_sim = 1-pairwise_distances(Description_vs_Invoice_matrix_cluster1.T, metric = "jaccard")
item_item_jac_sim = pd.DataFrame(item_item_jac_sim, index=Description_vs_Invoice_matrix_cluster1.columns, columns=Description_vs_Invoice_matrix_cluster1.columns)
item_item_jac_sim.head()
```
